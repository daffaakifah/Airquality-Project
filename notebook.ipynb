### Cleaning Data

```python
# Fungsi untuk menangani missing value
def handle_missing_values(data):
    # Mengisi missing value dengan rata-rata untuk kolom numerik
    for column in data.select_dtypes(include=['float64', 'int64']).columns:
        data[column].fillna(data[column].mean(), inplace=True)

    # Mengisi kolom kategori dengan nilai modus
    for column in data.select_dtypes(include=['object']).columns:
        data[column].fillna(data[column].mode()[0], inplace=True)

    return data
```

```python
# Menangani missing value untuk setiap dataset
data_Aotizhongxin = handle_missing_values(data_Aotizhongxin)
data_Changping = handle_missing_values(data_Changping)
data_Dingling = handle_missing_values(data_Dingling)
data_Dongsi = handle_missing_values(data_Dongsi)
data_Guanyuan = handle_missing_values(data_Guanyuan)
data_Gucheng = handle_missing_values(data_Gucheng)
data_Huairou = handle_missing_values(data_Huairou)
data_Nongzhanguan = handle_missing_values(data_Nongzhanguan)
data_Shunyi = handle_missing_values(data_Shunyi)
data_Tiantan = handle_missing_values(data_Tiantan)
data_Wanliu = handle_missing_values(data_Wanliu)
data_Wanshouxigong = handle_missing_values(data_Wanshouxigong)
```

```python
# Menampilkan jumlah missing value setelah penanganan
print("Jumlah missing value setelah data dihandling:")
print("Aotizhongxin:", data_Aotizhongxin.isna().sum().sum())
print("Changping:", data_Changping.isna().sum().sum())
print("Dingling:", data_Dingling.isna().sum().sum())
print("Dongsi:", data_Dongsi.isna().sum().sum())
print("Guanyuan:", data_Guanyuan.isna().sum().sum())
print("Gucheng:", data_Gucheng.isna().sum().sum())
print("Huairou:", data_Huairou.isna().sum().sum())
print("Nongzhanguan:", data_Nongzhanguan.isna().sum().sum())
print("Shunyi:", data_Shunyi.isna().sum().sum())
print("Tiantan:", data_Tiantan.isna().sum().sum())
print("Wanliu:", data_Wanliu.isna().sum().sum())
print("Wanshouxigong:", data_Wanshouxigong.isna().sum().sum())
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

Bagaimana distribusi nilai PM2.5 selama periode pengamatan?

```python
# Visualisasi
plt.figure(figsize=(12, 6))
sns.histplot(combined_data['PM2.5'].dropna(), bins=50, kde=True, color='blue')
plt.title('Distribusi Nilai PM2.5 Selama Periode Pengamatan')
plt.xlabel('PM2.5 (g/m)')
plt.ylabel('Frekuensi')
plt.grid()
plt.show()
```

### Pertanyaan 2:

Bagaimana tren rata-rata tahunan dari berbagai polutan (PM2.5, PM10, SO2, NO2, CO, O3 serta polutan apa yang berpengaruh paling signifikan dalam kualitas udara?

```python
# Mengonversi kolom tahun, bulan, dan hari menjadi datetime
combined_data['date'] = pd.to_datetime(combined_data[['year', 'month', 'day']])

# Memilih kolom yang relevan untuk analisis
relevant_columns = ['date', 'PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']
data_relevant = combined_data[relevant_columns].dropna()

# Menghitung rata-rata tahunan untuk setiap polutan
annual_avg = data_relevant.resample('Y', on='date').mean().reset_index()

# Membuat plot garis untuk menunjukkan tren rata-rata tahunan dari berbagai polutan
plt.figure(figsize=(14, 8))
sns.lineplot(data=annual_avg, x='date', y='PM2.5', label='PM2.5', marker='o')
sns.lineplot(data=annual_avg, x='date', y='PM10', label='PM10', marker='o')
sns.lineplot(data=annual_avg, x='date', y='SO2', label='SO2', marker='o')
sns.lineplot(data=annual_avg, x='date', y='NO2', label='NO2', marker='o')
sns.lineplot(data=annual_avg, x='date', y='CO', label='CO', marker='o')
sns.lineplot(data=annual_avg, x='date', y='O3', label='O3', marker='o')

plt.title('Tren Rata-rata Tahunan dari Berbagai Polutan Kualitas Udara')
plt.xlabel('Tahun')
plt.ylabel('Rata-rata Konsentrasi (g/m)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()
```

**Insight:**
- Distribusi nilai PM2.5 selama periode pengamatan terus mengalami penurunan frekuensi selama periode berlangsung
- Pengaruh polutan CO terlihat lebih signifikan dibanding polutan lainnya

## Conclusion

- Conclution pertanyaan 1: Selama periode pengamatan kualitas udara berlangsung, terlihat bahwa frekuensi PM2.5 mengalami penurunan. Hal ini dapat menjadi acuan untuk penanganan penurunan frekuensi PM2.5 di daerah lain. Bagaimana daerah setempat menangani hal tersebut atau faktor apa saja yang menjadi penyebebab berkurangnya kadar PM2.5 di seluruh daerah stasiun tersebut.
- Conclution pertanyaan 2: CO menjadi polutan dengan nilai yang terlihat signifikan, ini bisa menjadi bahan acuan mengenai fokus utama untuk mengurangi polutan CO dalam kadar udara tanpa mengabaikan polutan yang lainnya.
